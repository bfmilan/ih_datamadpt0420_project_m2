```python
import pandas as pd
from scipy.stats import t, norm, ttest_1samp, ttest_ind
```

```python
diamonds = pd.read_csv('../data/diamonds_train.csv')
```

```python
diamonds
```

Are our sample proces significantly different from 4,000

```python
test_results = ttest_1samp(diamonds['price'], 4000)
print('t suggests that the price might be smaller than 4000 EUR as it is', test_results.statistic)
print ('We can reject the null hypothesis as the p-value is', test_results.pvalue, 'Therefore, the avg. price is significantly different from 4000 EUR')
```
