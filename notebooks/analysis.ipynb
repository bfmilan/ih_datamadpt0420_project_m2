#### Challenge 1

You need to create a **Data Pipeline** that retrieves the following table:

| Country | Job Title      | Gender (1) | Quantity | Percentage |
|---------|----------------|------------|----------|------------|
| Spain   | Data Scientist | Male       | 25       | 5%         |
| Spain   | Data Scientist | Female     | 25       | 5%         |
| ...     | ...            | ...        | ...      | ...        |

```python
df_clean= df[['Country', 'Job Title', 'Gender', 'uuid']]
df_clean.head(2)
```

```python
df_final= df.groupby(['Country', 'Job Title', 'Gender'])['uuid'].count().reset_index().rename(columns = {'uuid':'Quantity'})
```

```python
country_filter = df_final['Country']=='Spain'
```

```python
df_result1 = df_final.loc[country_filter]
```

```python
df_result1['Percentage'] = df_final['Quantity']/df_final['Quantity'].sum()*1000
```
